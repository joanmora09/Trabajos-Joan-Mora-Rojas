<h4 Style = "color:#00008B"> 2.3 Imputacion simple de valores numericos </h4>

```python

import numpy as np
import pandas as pd

DfAutos["normalized"] = pd.to_numeric(DfAutos["normalized"], errors='coerce')
promNormalizadas = DfAutos["normalized"].mean()
DfAutos["normalized"].fillna(promNormalizadas, inplace=True)
DfAutos.head(5)
```

```python
promNormalizadas= DfAutos["bore"].astype("float").mean(axis=0)
DfAutos["bore"].replace(np.nan, promNormalizadas, inplace=True)
DfAutos.head(5)

promNormalizadas= DfAutos["stroke"].astype("float").mean(axis=0)
DfAutos["stroke"].replace(np.nan, promNormalizadas, inplace=True)
DfAutos.head(5)

promNormalizadas= DfAutos["peak"].astype("float").mean(axis=0)
DfAutos["peak"].replace(np.nan, promNormalizadas, inplace=True)
DfAutos.head(5)

promNormalizadas= DfAutos["horsepower"].astype("float").mean(axis=0)
DfAutos["horsepower"].replace(np.nan, promNormalizadas, inplace=True)
DfAutos.head(5)
```

<h5 Style = "color:#00008B"> 6.1. Configuracion de los graficos </h5>
<p> Cuantos autos existen segun fabricante?
    Cual fabricante tiene mayor precencia en el parque de autos? </p>

```python
#Remueve las lineas de los axis.
mpl.rcParams["axes.spines.bottom"] =True
mpl.rcParams["axes.spines.right"] =False
mpl.rcParams["axes.spines.left"] =False
mpl.rcParams["axes.spines.top"] =False
```

```python
gf=DfAutos.groupby(['make']).count()
gf=gf.iloc[:,:1]
gf=gf.sort_values('symboling', ascending= True)
gf
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(7, 6))  

eje_x = gf.index.tolist()
data = gf["symboling"].tolist()
bar_colors=['#FA6900','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5',
           '#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#69d2e7',]

ax.barh(eje_x, data, color=bar_colors)

ax.set_xlabel('Cant de carros')
ax.set_title('Carros segun fabricante')


plt.show()



#-------------------------CONFIGURACION DE LOS TITULOS, SUBTITULOS Y EJES DEL GFRAFICO----------------------------------------#

#FORMATO DEL TITULO:
title = "Grafico 1: Vehiculos en inventario"
plt.suptitle(title,
horizontalalignment='left',
x=0.001,
y=0.87,
transform=fig.transFigure,
fontsize= 20,
fontweight="bold",
fontfamily="serif",
fontname="Liberation Sans Narrow")

#FORMATO DEL SUBTITULO

subTitle = "Cantidad de vehiculos agrupados segun fabricante"
ax.set_title(subTitle,
       horizontalalignment= "left",
     x=0.001,
    y=0.80)

#EJE (X/Y)

etiqueta="cantidad de carros (ABS)"
ax.set_xlabel(etiqueta,
ha='left',
fontsize=10,
fontweight= 'regular',
fontfamily='serif')

ax.xaxis.set_label_coords(x=0, y=-0.07, transform=ax.transAxes)

#Formato tamaño de letra en ambos ejes
ax.tick_params(axis='both', labelsize=10)

#Configura la ubicacion, el titulo, subtitulo y grafico
plt.subplots_adjust(top=0.80)

plt.show()







```
