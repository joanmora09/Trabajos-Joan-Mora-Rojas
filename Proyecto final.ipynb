```python
gf = df_propiedades.groupby(['saleAmount']).count()
gf = gf.iloc[:, :1]
gf = gf.rename(columns={'salesRatio': 'count'})  # Renombrar la columna para mayor claridad
print(gf)
```

```python
# Realizar la agrupación
gf = df_propiedades.groupby(['saleAmount']).size()

# Convertir la serie resultante en un DataFrame
gf = gf.to_frame(name='count')

# Ordenar los resultados
gf = gf.sort_values('count', ascending=True)

# Crear el gráfico de barras
fig, ax = plt.subplots(figsize=(7, 6))

eje_x = gf.index.tolist()
data = gf["count"].tolist()
bar_colors = ['#FA6900', '#e5e5e5', '#e5e5e5', '#e5e5e5', '#e5e5e5', '#e5e5e5', '#e5e5e5', '#e5e5e5', '#e5e5e5', '#e5e5e5',
              '#e5e5e5', '#e5e5e5', '#e5e5e5', '#e5e5e5', '#e5e5e5', '#e5e5e5', '#e5e5e5', '#e5e5e5', '#e5e5e5', '#e5e5e5',
              '#e5e5e5', '#69d2e7']

ax.barh(eje_x, data, color=bar_colors)

ax.set_xlabel('cantidad de propiedades')
```
